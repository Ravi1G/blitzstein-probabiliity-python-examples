## Brief summary

### Change of variables in one dimension

Let $X$ be a continuous r.v. with PDF $f_X$, and let $Y = g(X)$, where $g$ is differentiable and strictly increasing (or strictly decreasing). Then the PDF of $Y$ is given by

\begin{equation}
f_Y(y) = f_X(x)\big|\frac{dx}{dy}\big|,
\end{equation}

where $x = g^{-1}(y)$. The support of $Y$ is all $g(x)$ with $x$ in the support of $X$.

### Change of variables

Let $\boldsymbol{X} = (X_1, ..., X_n)$ be a continuous random vector with joint PDF $\boldsymbol{X}$, and let $\boldsymbol{Y} = g(\boldsymbol{X})$ where $g$ is an invertible function from $\mathbb{R}^n$ to $\mathbb{R}^n$. (The domain of $g$ does not actually have to be all of $\mathbb{R}^n$, but it does need to be large enough to contain the support of $\boldsymbol{X}$, else $g(\boldsymbol{X})$ could be undefined!)

Let $\boldsymbol{y} = g(\boldsymbol{x})$, and suppose that all the partial derivatives $\frac{\partial x_i}{\partial y_j}$ exist and are continuous, so we can form the *Jacobian matrix*

\begin{equation}
\frac{\partial \boldsymbol{x}}{\partial \boldsymbol{y}} = 
\begin{bmatrix}
    \frac{\partial x_1}{\partial y_1} & \frac{\partial x_1}{\partial y_2} & ... & \frac{\partial x_1}{\partial y_n} \\ 
    ... &  &  & ... \\ 
    \frac{\partial x_n}{\partial y_1} & \frac{\partial x_n}{\partial y_2} & ... & \frac{\partial x_n}{\partial y_n}
\end{bmatrix}.
\end{equation}

Also assume that the determinant of the Jacobian matrix is never $0$. Then the joint PDF of $\boldsymbol{Y}$ is

\begin{equation}
f_{\boldsymbol{Y}}(\boldsymbol{y}) = f_{\boldsymbol{X}}(\boldsymbol{x})\big|\frac{\partial \boldsymbol{x}}{\partial \boldsymbol{y}}\big|,
\end{equation}

where the vertial bars say "take the absolute value of the determinant of $\partial \boldsymbol{x} / \partial \boldsymbol{y}$". As in the 1D case,

\begin{equation}
\big|\frac{\partial \boldsymbol{x}}{\partial \boldsymbol{y}}\big| = \big|\frac{\partial \boldsymbol{y}}{\partial \boldsymbol{x}}\big| ^{-1},
\end{equation}

so we can compute whichever of the two is easier, and then at the end express the joint PDF of $\boldsymbol{Y}$ as a function of $\boldsymbol{y}$.


### Convolution sums and integrals

If $X$ and $Y$ are independent discrete r.v.s, then the PMF of their sum $T = X + Y$ is

\begin{equation}
P(T=t) = \sum_x P(Y=t-x)P(X=x) \\
= \sum_y P(X=t-y)P(Y=y).
\end{equation}

If $X$ and $Y$ are independent continuous r.v.s, then the PDF of their sum $T = X + Y$ is

\begin{equation}
f_T(t) = \int_{-\infty}^{\infty} f_Y(t-x)f_X(x)dx \\
= \int_{-\infty}^{\infty} f_X(t-y)f_Y(y)dy.
\end{equation}


### Beta distribution

A r.v. $X$ is said to have the *Beta distribution* with parameters $a$ and $b$, $a > 0$ and $b > 0$, if its PDF is

\begin{equation}
f(x) = \frac{1} {\beta(a,b)} x^{a-1}(1-x)^{b-1},\ 0 < x < 1,
\end{equation}

where the constant $\beta(a, b)$ is chosen to make the PDF integrate to $1$. We write this as $X \sim Beta(a, b)$.

- Taking $a = b = 1$, the $Beta(1, 1)$ PDF is constant on (0, 1): $Unif(0,1)$.
- If $a < 1$ and $b < 1$, the PDF is U-shaped and opens upward. If $a > 1$ and $b > 1$, the PDF opens down.
- If $a = b$, the PDF is symmetric about $1/2$. If $a > b$, the PDF favors values larger than $1/2$; if $a < b$, the PDF favors values smaller than $1/2$.

By definition, the constant $\beta(a,b)$ satisfies 

\begin{equation}
\beta(a,b) = \int_0^1 x^{a-1}(1-x)^{b-1}dx.
\end{equation}

#### Beta-Binomial conjugacy

We have a coin that lands Heads with probability $p$, but we don't know what $p$ is. Our goal is to *infer* the value of $p$ after observing the outcomes of $n$ tosses of the coin. The larger that $n$ is, the more accurately we should be able to estimate $p$.

One major approach is *Bayesian inference*, which treats all unknown quantities as random variables. We would treat the unknown probability $p$ as a random variable and give $p$ a *prior* distribution. After the experiment is performed and the data are gathered, the prior distribution is updated using Bayes' rule; this yields the *posterior* distribution, which reflects our new beliefs about $p$.

let $p \sim Beta(a,b)$ for known constants $a$ and $b$, and let $X$ be the number of heads in $n$ tosses of the coin. Conditional on knowing the true value $p$, the tosses would just be independent Bernoulli trials with probability $p$ of success, so

\begin{equation}
X|p \sim Bin(n,p).
\end{equation}

After observing $k$ Heads, the posterior distribution of $p$ is

\begin{equation}
p|X = k \sim Beta(a+k, b+n-k).
\end{equation}

We have a Beta prior distribution on $p$ and data that are conditionally Binomial given $p$, then when going from prior to posterior, we don't leave the family of Beta distributions. We say that *the Beta is the conjugate prior of the Binomial*.

- The parameters $a$ and $b$ can be interpreted as the prior number of successes and failures, respectively.


### Gamma distribution

#### Gamma function

The *gamma function* $\Gamma$ is defined by

\begin{equation}
\Gamma(a) = \int_0^{\infty} x^ae^{-x}\frac{dx}{x},
\end{equation}

for real numbers $a > 0$.

- $\Gamma(a+1) = a\Gamma(a)$ for all $a > 0$. This follows from integration by parts:

\begin{equation}
\Gamma(a+1) = \int_0^{\infty} x^ae^{-x}dx = -x^ae^{-x}\big|_0^{\infty} + a\int_0^{\infty}x^{a-1}e^{-x}dx = 0 + a\Gamma(a).
\end{equation}

- $\Gamma(n) = (n-1)!$ if $n$ is a positive integer. This can be proved by induction.

#### Gamma distribution

A r.v. $Y$ is said to have the *Gamma distribution* with parameters $a$ and $\lambda$, $a > 0$ and $\lambda > 0$, if its PDF is 

\begin{equation}
f(y) = \frac{1}{\Gamma(a)} (\lambda y)^ae^{-\lambda y}\frac{1}{y},\ y > 0.
\end{equation}

We write $Y \sim Gamma(a, \lambda)$.

- $E(Y) = \frac{a}{\lambda}$
- $Var(Y) = \frac{a}{\lambda^2}$
- $E(Y^c) = \frac{1}{\lambda^c} \cdot \frac{\Gamma(a+c)}{\Gamma(a)},\ c > -a$

#### Gamma and Exponentials

Let $X_1, ..., X_n$ be i.i.d. $Expo(\lambda)$. Then

\begin{equation}
X_1 + ... + X_n \sim Gamma(n, \lambda).
\end{equation}

- In a Poisson process of rate $\lambda$, the total waiting time for $n$ arrivals is distributed $Gamma(n,\lambda)$. Thus the Gamma is the continuous analog of the Negative Binomial distribution.

#### Gamma-Poisson conjugacy

In Blotchville, buses arrive at a certain bus stop according to a Poisson process with rate $\lambda$ buses per hour, where $\lambda$ is unknown. Based on his very memorable adventures in Blotchville, Fred quantifies his uncertainty about $\lambda$ using the prior $\lambda \sim Gamma(r_0, b_0)$, where $r_0$ and $b_0$ are known, positive constants with $r_0$ an integer.

By the definition of Poisson process, conditional on knowing the true rate $\lambda$, the number of buses in an interval of length $t$ is distributed $Pois(\lambda t)$. In other words, 

\begin{equation}
\lambda \sim Gamma(r_0, b_0) \\
Y|\lambda \sim Pois(\lambda t).
\end{equation}

After observing $y$ buses, the posterior distribution of $\lambda$ is 

\begin{equation}
\lambda | Y = y \sim Gamma(r_0+y, b_0+t).
\end{equation}

- The parameters $r_0$ and $b_0$ can be interpreted as the prior number of observed successes and the total waiting time for those successes, respectively.


### Order statistics

For r.v.s $X_1, X_2, ..., X_n$, the *order statistics* are the random variables $X_{(1)}, ..., X_{(n)}$, where

\begin{equation}
X_{(1)} = min(X_1, ..., X_n), \\
X_{(2)}\ \text{is the second-smallest of}\ X_1, ..., X_n, \\
... \\
X_{(n-1)}\ \text{is the second-largest of}\ X_1, ..., X_n, \\
X_{(n)} = max(X_1, ..., X_n).
\end{equation}

Note that $X_{(1)} \leq X_{(2)} \leq ... \leq X_{(n)}$ by definition. We call $X_{(j)}$ the *jth order statistic*. If $n$ is odd, $X_{((n+1)/2}$ is called the *sample median* of $X_1, ..., X_n$.

- The order statistics $X_{(1)}, ..., X_{(n)}$ are random variables, and each $X_{(j)}$ is a function of $X_1, ..., X_n$. Even if the original r.v.s are independent, the order statistics are *dependent*.

#### CDF of order statistic

Let $X_1, ..., X_n$ be i.i.d. continuous r.v.s with CDF $F$. Then the CDF of the $j$th order statistic $X_{(j)}$ is

\begin{equation}
P(X_{(j)} \leq x) = \sum_{k=j}^n {{n \choose k}} F(x)^k(1-F(x))^{n-k}.
\end{equation}

#### PDF of order statistic

Let $X_1, ..., X_n$ be i.i.d. continuous r.v.s with CDF $F$ and PDF $f$. Then the marginal PDF of the $j$th order statistic $X_{(j)}$ is

\begin{equation}
f_{X_{(j)}}(x) = n{{n-1 \choose j-1}}f(x)F(x)^{j-1}(1-F(x))^{n-j}.
\end{equation}


### Beta-Gamma connections

If $X \sim Gamma(a, \lambda)$, $Y \sim Gamma(b, \lambda)$, and $X$ and $Y$ are independent, then $X + Y \sim Gamma(a+b, \lambda)$, $\frac{X}{X+Y} \sim Beta(a,b)$, and $X+Y$ and $\frac{X}{X+Y}$ are independent.

### Bayes' billiards

In the Bayes' billiards story, we have $n$ white balls and $1$ gray ball, throw them onto the unit interval completely at random, and count the number of white balls to the left of the gray ball. Letting $p$ be the position of the gray ball and $X$ be the number of white balls to the left of the gray ball, we have

\begin{equation}
p \sim Unif(0, 1) \\
X|p \sim Bin(n, p).
\end{equation}

By performing this experiment a large number of times, we can verify the results we derived about the marginal PMF of $X$ and the posterior PDF of $p$ given $X=x$.
